Las computadoras funcionan de principio a fin gracias a que hacen operaciones matemáticas. Dentro de python también podemos hacer uso de estas operaciones aritmeticas para calcular cosas.

<h2><b>Operadores Aritmeticos</b></h2>


```python
print('Suma:', 1 + 2)
print('Resta:', 2 - 1)
print('Multiplicación:', 2 * 3)
print('Division:', 4 / 2) #La división en Python asigna automáticamente números flotantes
print('Division:', 7 / 2)
print('Division sin decimales:', 7 // 2)
print('Modulo:', 3 % 2)
print('Exponente:', 2 ** 3)
```

Podemos también unir palabras haciendo uso de operadores

```python
print("Hola", "Python")
print("Hola" + "Python")
print("Hola" + str(5))
print("Hola " * 5)
```

```python
##Estas funciones nos facilitan mucho ciertos trabajos como por ejemplo
for i in range(11):
    print("*" * i)
```

Además podemos hacer uso de estos operadores junto a las variables.

```python
pizza = 20
refresco = 15

compra = pizza + refresco
propina = compra * 0.2
total = compra + propina

print("Usted debe pagar:", compra ,"De propina usted debe:", propina ,"Su total es:", total)
```

<h2><b>Operadores de comparación</b></h2>

```python
print('Igual a:', 2 == 2) #TRUE
print('Igual a:', 2 == 2.00) #VILLA DICE QUE NO
print('Igual a:', 2 == 1.99999) #FALSE
```

```python
print('Distinto de:', 2 != 2) #FALSE
print('Distinto de:', 2 != 3) #TRUE
```

```python
print('Mayor que:', 2 > 3) #FALSE
print('Menor que:', 2 < 3) #TRUE
print('Mayor o igual que:',4 >= 3) #TRUE 
print('Mayor o igual que:',3 >= 3) #TRUE
print('Menor o igual que', 3 <= 4) #TRUE
print('Menor o igual que:',4 <= 4) #TRUE
```

También podemos hacer uso de estos operadores junto a textos


```python
print(len('mango') == len('melón'))
print(len('mango') != len('melón'))
print(len('mango') < len('melón'))
print(len('mango') > len('melón'))
```

<h3><b>Operadores de comparación con texto</b></h3>

* ***is***  Regresa <b>True</b> si ambas variables son del mismo tipo.
* ***is not***  Regresa <b>True</b> si ambas variables NO son del mismo tipo.
* ***in*** Regresa <b>True</b> si el valor inicial se encuentra en el valor secundario
* ***not in*** Regresa <b>True</b> si el valor inicial NO se encuentra en el valor secundario

```python
print('1 is 1', 1 is 1)                   # Verdadero - El tipo de dato es el mismo
print('1 is not 2', 1 is not 2)           # True - Aunque son el mismo tipo de dato, no son de valor similar
print('A in Alejandro', 'A' in 'Alejandro') # True - A se encuentra en el string
print('L in Alejandro', 'L' in 'Alejandro') # False - No se encuentra una L mayúscula
print('coding' in 'coding for all') # True - porque la palabra se encuentra en la oración
print('a in an:', 'a' in 'an')      # True - Porque se encuentra la letra en la oración
print('4 is 2 ** 2:', 4 is 2 ** 2)   # True
```

<h2><b>Operadores Lógicos</b></h2>

* ***and***  Regresa <b>True</b> si ambas condiciones se cumplen
* ***or***  Regresa <b>True</b> si una o ambas condiciones se cumplen
* ***not*** Regresa <b>True</b> si el valor es False

```python
print(3 > 2 and 4 > 3) # True - porque ambas condiciones se cumplen
print(3 > 2 and 4 < 3) # False - porque la segunda condición es falsa
print(3 < 2 and 4 < 3) # False - porque ambas condiciones son falsas
print('True and True: ', True and True)

print(3 > 2 or 4 > 3)  # True - ambas condiciones se cumplen
print(3 > 2 or 4 < 3)  # True - una de las dos condiciones se cumple
print(3 < 2 or 4 < 3)  # False - ambas condiciones son falsas
print('True or False:', True or False)

print(not 3 > 2)     # False - la afirmación es verdadera, entonces regresa lo contrario
print(not True)      # False - Negación, el operador muestra el contrario
print(not False)     # True 
print(not not True)  # True
print(not not False) # False
```
